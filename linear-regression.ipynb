```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# House size (in 1000 sq ft)
x = np.array([0.5, 1.0, 1.5, 2.0, 2.5, 3.0])

# Price (in lakhs)
y = np.array([50, 80, 110, 130, 150, 180])

m = len(x)

plt.scatter(x, y)
plt.title("House Prices vs Size")
plt.xlabel("Size (1000 sq ft)")
plt.ylabel("Price (Lakhs)")
plt.show()
```

```python
def predict(x, w, b):
    return w * x + b
```

```python
def compute_cost(x, y, w, b):
    m = len(x)
    cost = 0

    for i in range(m):
        y_pred = w * x[i] + b
        error = y_pred - y[i]
        cost += error ** 2
    return cost / (2 * m)
```

```python
def gradient_descent(x, y, w, b, alpha, iterations):
    m = len(x)

    cost_history = []
    w_history = []

    for iter in range(iterations):
        dj_dw = 0
        dj_db = 0

        #calculate Gradients
        for i in range(m):
            y_pred = w * x[i] + b
            error = y_pred - y[i]

            dj_dw += error * x[i]
            dj_db += error

        dj_dw /= m
        dj_db /= m


        #update params
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        #save history
        cost = compute_cost(x, y, w, b)
        cost_history.append(cost)
        w_history.append(w)

        #print progress every 100 steps
        if iter % 1000 == 0:
            print(f"iteration{iter}: Cost={cost: .2f},w={w:.2f}, b={b:.2f}")

    return  w, b, cost_history, w_history


           
```

```python
w = 0
b = 0
alpha = 0.01
iterations = 10000

w, b, cost_history, w_history = gradient_descent(x, y, w, b, alpha, iterations)

print("\nFinal values:")
print("w =", w)
print("b =", b)
```

```python

y_pred = predict(x, w, b)


plt.scatter(x, y, label="Actual Data")
plt.plot(x, y_pred, color='red', label="Regression Line")
plt.xlabel("Size")
plt.ylabel("Price")
plt.legend()
plt.show()
```
