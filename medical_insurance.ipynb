```python
X = df.drop(columns=["charges"])
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
X_encoded
```

```python
X_encoded.shape
```

```python
X_encoded.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_encoded_train, X_encoded_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

```python
X_encoded_train.shape
```

```python
X_encoded_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_encoded_train,y_train)
```

```python
y_pred = model.predict(X_encoded_test)
y_pred
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
r2_score(y_test, y_pred)
```

```python
root_mean_squared_error(y_test, y_pred)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.title("Actual vs Predicted Charges")

# ideal line
min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], color="red", linestyle="--")

plt.show()
```

The actual vs predicted plot shows that the model captures the overall trend, especially for lower and moderate insurance charges. However, it tends to underpredict higher charges. The residual plot is not randomly scattered around zero and shows visible patterns, indicating that the simple linear regression model does not fully capture the underlying relationships in the data.

```python
residuals = y_test - y_pred

plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, alpha=0.7)
plt.axhline(y=0, color="red", linestyle="--")
plt.xlabel("Predicted Charges")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```
